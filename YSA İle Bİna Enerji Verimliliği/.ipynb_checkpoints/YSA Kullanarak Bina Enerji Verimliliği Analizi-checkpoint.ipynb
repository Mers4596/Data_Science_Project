# Eğitim Ve Test Veri Seti Ayırma 

```python
data_set.head()
```

```python
from sklearn.model_selection import train_test_split
X = data_set.iloc[:,0:8]
y = data_set.iloc[:,8:10]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
print("X_train of data set: \n", X_train , "\n \n \n ", "X_test of data set: \n", X_test, "\n \n \n"
     ,"y_train of data set: \n", y_train, "\n \n \n ", " y_test of data set: \n", y_test)
```

# Yapay Sinir Ağı İçin Array Haline Getirelim 

```python
import numpy as np 
X_train_scaled = np.array(X_train)
X_test_scaled = np.array(X_test)

y_train_scaled = np.array(y_train)
y_test_scaled = np.array(y_test)

y_train_scaled = (y_train_scaled[:,0], y_train_scaled[:,1])
```

```python
print("X_test, X_train, y_test, y_train  Array Haline Getirlimiş Halleri \n \n")
print("X_train_scaled of data set: \n", X_train_scaled , "\n \n \n ", "X_test_scaled of data set: \n", X_test_scaled, "\n \n \n"
     ,"y_train_scaled of data set: \n", y_train_scaled, "\n \n \n ", " y_test_scaled of data set: \n", y_test_scaled)
```

# Yapay Sinir Ağını Oluşturalım

```python
from tensorflow.keras.layers import Input, Dense, Dropout
from tensorflow.keras.models import Model

# Giriş katmanı, verinin özellik sayısı kadar giriş alıyor
input_layer = Input(shape=(X_train_scaled.shape[1],), name = 'Input_Layer')

# İlk gizli katman: 128 nöron, ReLU aktivasyon
common_path = Dense(units=128 , activation = 'tanh', name = 'First_Dense')(input_layer)

# %30 dropout (overfitting’i önlemek için nöronların %30’unu kapatır)
common_path = Dropout(0.3)(common_path)

# İkinci gizli katman: 128 nöron, ReLU aktivasyon
common_path = Dense(units = 128, activation='tanh', name = 'Second_Dense')(common_path)

# Yine %30 dropout
common_path = Dropout(0.3)(common_path)

# 1. Çıkış: direkt olarak ortak yolun sonunda tek nöronlu bir çıkış
first_output = Dense(units=1, name='First_Output_Last_Layer')(common_path) # Evet, Hayır 

# 2. Çıkış için ek gizli katman: 64 nöron, ReLU aktivasyon
second_output_path = Dense(units=64, activation='tanh', name='Second_Output_First_Dense')(common_path)

# %30 dropout
second_output_path = Dropout(0.3)(second_output_path)

# 2. Çıkış: tek nöronlu katman
second_output = Dense(units=1, name='Second_Output_Last_Layer')(second_output_path) # daha ayrıntılu öğreniyor ve dereceyi ölçüyor

# Model tanımı: tek giriş, iki farklı çıkış
model = Model(inputs=input_layer, outputs=[first_output, second_output])

# Modelin özetini yazdır
print(model.summary())
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping


optimizer = tf.keras.optimizers.Adam(learning_rate=0.0001, clipnorm=1.0) 


##!!!!!!!!!!! Loss(Kayıp Değerler) !!!!!!!!!!! eğitim için kullanılırken Metrics Performans İçin kullanılıyor

model.compile(optimizer=optimizer, # compile kısmına optimizer kısmnını yukarıda tanımladığımız optimzer olarak atyıyoruz
              loss # Kayıp Değerlerin Hesaplandığı kısım bu kayıp değerleri hesaplarkende mse kullanıyor
              ={'First_Output_Last_Layer': 'mse', # Hata çözme yöntemi(En yaygın kullanılan yaptığı hata kadar ceza ver anlamına gelir)
                    'Second_Output_Last_Layer': 'mse'},
              metrics # Modelin performansını insan gözüyle takip etmemize yarayan bir ölçüm aracı 
              ={'First_Output_Last_Layer': tf.keras.metrics.RootMeanSquaredError(),
                       'Second_Output_Last_Layer': tf.keras.metrics.RootMeanSquaredError()})


# EarlyStopping Bir koşucu gibi düşünelim bu koşunun yorulmasını ve performansının düşmesini engellemek için kullanılır
earlyStopping_callback = EarlyStopping(monitor='val_loss', # İzlenecek değeri belli eder
                                       min_delta=0, # Bu, en az ne kadar iyileşmenin bir "iyileşme" olarak kabul edileceğini belirtir. Burada 0 olduğu için, en ufak bir iyileşme bile sabır süresini sıfırlar.
                                       patience=10, # Bu, modelin "sabır süresidir." Validasyon kaybında iyileşme görülmediği halde model 10 dönem (epoch) daha eğitime devam eder. Eğer 10 dönem sonra hâlâ bir iyileşme yoksa, eğitim durdurulur.
                                       verbose=1) # Bu, eğitim durduğunda ekrana bir mesaj yazdırılmasını sağlar. Bu sayede, eğitimin neden durduğunu görebilirsiniz.
```

```python
history = model.fit(x= X_train_scaled, 
                    y = y_train_scaled, 
                    epochs=500,
                    verbose=0,
                    batch_size=10, 
                    validation_split=0.3, 
                    callbacks=earlyStopping_callback)
y_predict = np.array(model.predict(X_test_scaled))

from sklearn.metrics import r2_score
print("İlk Çıkışın r2 Socre: ", r2_score(y_test_scaled[:,0], y_predict[0,:].flatten()))
print("İkinci Çıkışın r2 Socre: ", r2_score(y_test_scaled[:,1], y_predict[1,:].flatten()))
```
