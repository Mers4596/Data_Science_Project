```python
import pandas as pd
import numpy as np 

train_data = pd.read_csv("C:/Users/Mehmet Ersolak/Documents/GitHub/Data_Science_Project/EKG Veri Setinin Analizi/mitbih_train.csv")
print(len(train_data))

test_data = pd.read_csv("C:/Users/Mehmet Ersolak/Documents/GitHub/Data_Science_Project/EKG Veri Setinin Analizi/mitbih_test.csv")
print(len(test_data))
```

```python
X_train = np.array(train_data)[:,:187]
y_train = np.array(train_data)[:,187]

print(f"Lenght of X_train: {len(X_train)}, Lenght of y_train: {len(y_train)}")

X_test = np.array(test_data)[:,:187]
y_test = np.array(test_data)[:,187]

print(f"Lenght of X_test: {len(X_test)}, Lenght of y_test: {len(y_test)}")
```

```python
from sklearn.svm import SVC
model = SVC(kernel="linear", C = 0.1)
model.fit(X_train, y_train)
```

## model = SVC(kernel = "linear", C = 0.1)
kernel(Çekirdek): 
kernel sayesinde doğrusal olarak ayrılmayan verileri farklı bir iki boyutlu uzaya taşıyarak "linear" sayesinde doğrusal olarak ayırmak için bir nokta bulmaya çalışır.

C(Denge Parametresi):
Eğitim verilerindeki her bir noktayı kusursuzca ayırmak ile daha genel ve yeni verilerde de iyi çalışacak bir model oluşturmak arasında bir denge kurmaktır. $C$ parametresi işte bu dengeyi ayarlar.

Yüksek $C$ Değeri (örn: 10, 100): 
Modelin eğitim verilerindeki hatalara karşı çok hassas olmasını sağlar.

Düşük $C$ Değeri (örn: 0.1, 0.01): Modelin daha "hoşgörülü" olmasını sağlar. Model, bazı eğitim verilerini yanlış sınıflandırma pahasına daha basit ve daha genel bir karar sınırı bulmaya çalışır.

+Bu iki satir kodla yaptiginiz sey sudur:+

Sınıflandırma yapmak amacıyla, verileri doğrusal bir çizgi (veya hiperdüzlem) ile ayırmaya çalışacak ve aşırı öğrenmeyi önlemek için hatalara karşı bir miktar tolerans gösterecek (yumuşak marjlı) bir Destek Vektör Sınıflandırıcısı modeli oluşturdunuz.

```python
from sklearn.metrics import accuracy_score

y_predict_train = model.predict(X_train)
accuracy_train = accuracy_score(y_predict_train, y_train)
print("Train Data Accuracy: ", accuracy_train)

y_predict_test = model.predict(X_test)
accuracy_test = accuracy_score(y_predict_test, y_test)
print("Test Data Accuracy: ", accuracy_test)

```

```python
from sklearn.metrics import confusion_matrix 
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_predict_test, y_test)
index = ['N','S','V','F','Q']
columns = ['N','S','V','F','Q']

confusion_matrix = pd.DataFrame(cm, columns, index)

print("cm: \n",cm, "\n")
plt.figure(figsize=(10,6))
plt.title("Confusion_Matrix")
sns.heatmap(confusion_matrix, annot = True, fmt='d', cmap = "YlGnBu");


```
