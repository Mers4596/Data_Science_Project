## tarin test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=15)
```

## Redundancy, Multicollinearity, Overfitting

```python
def correlation_for_droping(df, threshold):
    columns_to_drop = set()
    corr = df.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i, j]) > threshold:
                columns_to_drop.add(corr.columns[i])
    return columns_to_drop
```

```python
columns_droping = correlation_for_droping(X_train, 0.85)
columns_droping
```

```python
X_train = X_train.drop(columns_droping, axis=1)
X_test = X_test.drop(columns_droping, axis=1)
```

```python
X_train.drop(['day','month','year'], axis=1, inplace=True)
X_test.drop(['day','month','year'], axis=1, inplace=True)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler 
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.subplots(figsize = (15,5))

plt.subplot(1,2,1)
sns.boxplot(data = X_train)
plt.title('X_train')

plt.subplot(1,2,2)
sns.boxplot(data = X_train_scaled)
plt.title('X_train_scaled')

plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
y_pred = linear.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)

print('Mean Squared Error: ', mse)
print('Mean Absolute Error: ', mae)
print('R2 Score: ', score)

plt.scatter(y_test, y_pred)
plt.show()
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)

print('Mean Squared Error: ', mse)
print('Mean Absolute Error: ', mae)
print('R2 Score: ', score)

plt.scatter(y_test, y_pred)
plt.show()
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)

print('Mean Squared Error: ', mse)
print('Mean Absolute Error: ', mae)
print('R2 Score: ', score)

plt.scatter(y_test, y_pred)
plt.show()
```

```python
from sklearn.linear_model import ElasticNet
elastic = ElasticNet()
elastic.fit(X_train_scaled, y_train)
y_pred = elastic.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)

print('Mean Squared Error: ', mse)
print('Mean Absolute Error: ', mae)
print('R2 Score: ', score)

plt.scatter(y_test, y_pred)
plt.show()
```

## Lasso Cross Validation

```python
from sklearn.linear_model import LassoCV
```

```python
lassocv = LassoCV(cv = 5)
lassocv.fit(X_train_scaled, y_train)
y_pred = lassocv.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
score = r2_score(y_test, y_pred)

print('Mean Squared Error: ', mse)
print('Mean Absolute Error: ', mae)
print('R2 Score: ', score)

plt.scatter(y_test, y_pred)
plt.show()
```

```python
print("Normal Lasso Model's Alpha Value ---> ", lasso.alpha)
print("Cross Validation Lasso Model's Alpha Value ---> ", lassocv.alpha_)
```
